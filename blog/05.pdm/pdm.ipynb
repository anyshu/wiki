```python
from math import *
from matplotlib.pyplot import *


def dist(a, b):
    return abs(a - b)


def pdm(a):
    lst = []
    plus, neg = 1, -1
    e = 0
    for i in a:
        e += i
        d0, d1 = dist(e, plus), dist(e, neg)
        o = plus if d0 < d1 else neg # minimal difference
        e -= o # error diffuse
        lst.append(o)
    return lst

def showPdm(a):
    b = pdm(a)
    b = pdm(a)
    subplot(2, 1, 1)
    plot(a)
    subplot(2, 1, 2)
    plot(b)
    ylim(-1, 1)
    show()
a = [sin(i/80.) for i in range(400)]
showPdm(a)
a = [sin(400./(i + 1)) for i in range(400)]
showPdm(a)
```
