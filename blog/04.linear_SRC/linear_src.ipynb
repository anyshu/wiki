```python
from matplotlib.pyplot import *
from math import *


def src(m, n, xs):
    def inter(i):
        j, k = (i * m) // n, ((i * m) % n) / n
        return xs[j] * (1 - k) + xs[j + 1] * k
    return [inter(i) for i in range((len(xs) - 1) * n // m)]

x = [i / 2. for i in range(6 * 3 * 2)]
y = [sin(i) for i in x]

subplot(3, 1, 1)
plot(y, "o")
subplot(3, 1, 2)
z2 = src(7, 5, y)
plot(z2, "o")
subplot(3, 1, 3)
z3 = src(5, 7, y)
plot(z3, "o")
show()
```
